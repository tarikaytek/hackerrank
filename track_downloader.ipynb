```python
from pytube import YouTube
from google.colab import drive
from tqdm import tqdm
import os
import pandas as pd
```

```python
#google drive connection
drive.mount('/content/drive')
```

```python
def download_youtube_audio(video_id, isrc, output_path):
  """
  download low quality mp3 file of given song

  :video_id: universal id of the song
  :isrc: isrc code of the song
  :output_path: where 2 download the track
  """
    try:
        video_url = f'https://www.youtube.com/watch?v={video_id}'
        yt = YouTube(video_url)

        #configure quality. filter it to only audio and lowest quality.
        audio_stream = yt.streams.filter(only_audio=True).first()
        audio_stream.download(output_path, filename=f'{isrc}.mp3')

        print(f"ok: {isrc}")

    except Exception as e:
        print(f"error downloading {isrc}: {str(e)}")
```

```python
csv_file_path = '/content/drive/MyDrive/ColabNotebooks/scouting_track_subset.csv'
output_folder = '/content/drive/MyDrive/tracks'

# null drop ve duplicate drop
df = pd.read_csv(csv_file_path).dropna(subset=['video_id', 'isrc'])
df = df.drop_duplicates(subset=['isrc'])

#tqdm for progress bar
with tqdm(total=len(df)) as pbar:
    #for each row of read csv file
    for index, row in df.iterrows():
        video_id = row['video_id']
        isrc = row['isrc']

        #check if it is downloaded before
        #comment out if you start downloading from zero
        audio_file_path = os.path.join(output_folder, f'{isrc}.mp3')
        if os.path.exists(audio_file_path):
            print(f"Already downloaded: {isrc}")
            pbar.update(1)
            continue

        #call the download function
        download_youtube_audio(video_id, isrc, output_folder)
        pbar.update(1)

#kill the progress bar
pbar.close()
```
